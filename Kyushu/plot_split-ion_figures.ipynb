```python
import os
import numpy as np
import matplotlib.pyplot as plt
from cartopy import crs as ccrs
from mintpy.utils import readfile, utils as ut, plot as pp
from mintpy.cli import view, tsview, plot_network, plot_transection
from mintpy.view import prep_slice, plot_slice
import h5py
```

```python
ionStack_path = '../inputs/ionStack.h5'
with h5py.File(ionStack_path, 'r') as file:
    date = file['date']
    date = date[()]
ion_date_list = np.char.decode(date)
ion_date_list = [f'{start_date}_{end_date}' for start_date, end_date in ion_date_list]
```

```python
for i in range(len(ion_date_list)):
    command = f"../inputs/ionStack.h5 -n {i} --wrap -c cmy --nodisplay --noverbose --noreference --title {ion_date_list[i]}_ion -o {ion_date_list[i]}_ion.png "
    view.main(command.split())
```

```python
ifgramStack_path = '../inputs/ifgramStack.h5'
with h5py.File(ifgramStack_path, 'r') as file:
    date = file['date']
    date = date[()]
wrap_date_list = np.char.decode(date)
wrap_date_list = [f'{start_date}_{end_date}' for start_date, end_date in wrap_date_list]
```

```python
if wrap_date_list == ion_date_list:
    print("The two lists are exactly the same.")
else:
    print("The two lists are not the same.")
```

```python
for i in range(len(wrap_date_list)):
    command = f"../inputs/ifgramStack.h5 -n {i} --wrap -c cmy --nodisplay --noverbose --noreference --title {wrap_date_list[i]}_wrap -o {wrap_date_list[i]}_wrap.png "
    view.main(command.split())
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
import matplotlib.image as mpimg

for i in range(len(wrap_date_list)):
    print(f'Progress: {i} / {len(wrap_date_list)} ')
    # 读取图片
    img1 = mpimg.imread(f'{wrap_date_list[i]}_wrap.png')
    img2 = mpimg.imread(f'{ion_date_list[i]}_ion.png')

    # 创建图像
    fig, axes = plt.subplots(1, 2, figsize=(10, 6))

    # 在第一个子图上显示图像
    axes[0].imshow(img1)
    axes[0].axis('off')  # 不显示刻度

    # 在第二个子图上显示图像
    axes[1].imshow(img2)
    axes[1].axis('off')

    # 调整子图之间的空白并显示图像
    plt.tight_layout()
    
     # 保存图像，文件名为 ion_date_list[i].png，背景透明
    plt.savefig(f'{ion_date_list[i]}.png', transparent=True)
    plt.show()
```
