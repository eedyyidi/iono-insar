```python
import os
import datetime as dt
import numpy as np
from scipy import stats
from matplotlib import pyplot as plt, ticker
import cartopy.crs as ccrs
import cartopy.feature as cfeature
# 保留 mintpy 相关导入，若未使用可考虑移除
from mintpy.objects import timeseries
from mintpy.utils import ptime, readfile, utils as ut, plot as pp
from mintpy import tsview

name = 'usa'
lalo = [40, -118]

lat, lon = lalo
proj = ccrs.NearsidePerspective(central_longitude=lon, central_latitude=lat, satellite_height=600e3)
fig, ax = plt.subplots(nrows=1, ncols=1, figsize=[2, 2], subplot_kw=dict(projection=proj))
ax.add_feature(cfeature.LAND, edgecolor='none', facecolor='0.9')
ax.add_feature(cfeature.BORDERS, lw=0.5)
ax.add_feature(cfeature.STATES, linestyle='-', linewidth=0.5, edgecolor='gray')
ax.coastlines(resolution='50m', lw=0.5, alpha=1)

x1 = [-122.2165, -118.3796, -115.7264, -118.9552, -122.2165]
y1 = [47.3063, 32.0225, 32.4059, 47.689, 47.3063]
x2 = [-115.295044, -113.469086, -116.108665, -118.256187, -115.295044]
y2 = [40.520218, 31.378164, 30.970022, 40.124332, 40.520218]

# 指定数据投影为 PlateCarree
ax.plot(x1, y1, color='#5BAEFF', linewidth=1, transform=ccrs.PlateCarree())
ax.fill(x1, y1, color='#5BAEFF', alpha=0.6, transform=ccrs.PlateCarree())
ax.plot(x2, y2, color='#5BAEFF', linewidth=1, transform=ccrs.PlateCarree())
ax.fill(x2, y2, color='#5BAEFF', alpha=0.6, transform=ccrs.PlateCarree())

ax.set_global()

# 第一个箭头的参数
arrow1_length = 0.3  # 第一个箭头长度
arrow1_angle = 12  # 第一个箭头角度，单位为度
arrow1_angle_rad = np.radians(arrow1_angle)
arrow1_dx = arrow1_length * np.cos(arrow1_angle_rad)
arrow1_dy = arrow1_length * np.sin(arrow1_angle_rad)

# 第一个箭头的左下角坐标（使用轴坐标）
arrow1_x0, arrow1_y0 = 0.15, 0.2

# 添加第一个箭头
ax.annotate('', xy=(arrow1_x0 + arrow1_dx, arrow1_y0 + arrow1_dy), xytext=(arrow1_x0, arrow1_y0),
            xycoords='axes fraction', textcoords='axes fraction',
            arrowprops=dict(facecolor='black', edgecolor='black', shrink=0.05, width=1, headwidth=7))

# 第二个箭头的参数
arrow2_length = 0.5 # 第二个箭头长度
arrow2_angle = arrow1_angle + 90  # 第二个箭头角度，与第一个垂直
arrow2_angle_rad = np.radians(arrow2_angle)
arrow2_dx = arrow2_length * np.cos(arrow2_angle_rad)
arrow2_dy = arrow2_length * np.sin(arrow2_angle_rad)

# 第二个箭头的左下角坐标（使用轴坐标）
arrow2_x0, arrow2_y0 = 0.16, 0.17

# 添加第二个箭头
ax.annotate('', xy=(arrow2_x0 + arrow2_dx, arrow2_y0 + arrow2_dy), xytext=(arrow2_x0, arrow2_y0),
            xycoords='axes fraction', textcoords='axes fraction',
            arrowprops=dict(facecolor='black', edgecolor='black', shrink=0.05, width=1, headwidth=7))

# output
out_fig = f'loc_{name}.pdf'
print('save figure to file:', out_fig)
fig.savefig(out_fig, bbox_inches='tight', transparent=True, dpi=600)
plt.show()
```
